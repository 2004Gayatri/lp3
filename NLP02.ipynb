```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import TreebankWordTokenizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
```

```python
documents = [
    "Natural language processing is interesting",
    "Machine learning and NLP are closely related",
    "AI is transforming the world",
    "NLP includes tokenization and embedding techniques"
]
```

```python
# Bag of Words - Count Occurrence

count_vectorizer = CountVectorizer()

X_count = count_vectorizer.fit_transform(documents)

print("Count Occurrence Matrix:\n")
print(X_count.toarray())

print("\nFeature Names:\n")
print(count_vectorizer.get_feature_names_out())
```

```python
# Normalized Count Occurrence

X_norm = X_count.toarray().astype(float)

X_norm = X_norm / X_norm.sum(axis=1, keepdims=True)

print("Normalized Count Matrix:\n")
print(X_norm)
```

```python
# TF-IDF Representation

tfidf_vectorizer = TfidfVectorizer()

X_tfidf = tfidf_vectorizer.fit_transform(documents)

print("TF-IDF Matrix:\n")
print(X_tfidf.toarray())
```

```python
# Tokenization using Treebank Tokenizer

tokenizer = TreebankWordTokenizer()

tokenized_docs = [
    tokenizer.tokenize(doc.lower())
    for doc in documents
]

print(tokenized_docs)
```

```python
# Create Word2Vec Model

model = Word2Vec(
    sentences=tokenized_docs,
    vector_size=100,
    window=5,
    min_count=1,
    workers=1
)

# Print word embedding for 'nlp'

print("Word Embedding for 'nlp':\n")
print(model.wv['nlp'])
```

```python
# Print all word embeddings

print("All Word Embeddings:\n")

for word in model.wv.index_to_key:
    print(word, "->", model.wv[word])
```
