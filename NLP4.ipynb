```python
import torch
import torch.nn as nn
```

```python
# Simple Transformer Model
class SimpleTransformer(nn.Module):

    def __init__(self):
        super(SimpleTransformer, self).__init__()

        # Transformer Layer
        self.transformer = nn.Transformer(
            d_model=4,               # Embedding size
            nhead=2,                 # Number of attention heads
            num_encoder_layers=1,    # Encoder layers
            num_decoder_layers=1     # Decoder layers
        )

    def forward(self, src, tgt):

        # Pass source and target through transformer
        output = self.transformer(src, tgt)

        return output
```

```python
# Create model object
model = SimpleTransformer()

print("Transformer Model Created Successfully")
```

```python
# Random source input
src = torch.rand((5, 1, 4))

# Random target input
tgt = torch.rand((5, 1, 4))

print("Source Shape :", src.shape)
print("Target Shape :", tgt.shape)
```

```python
# Run transformer model
output = model(src, tgt)

# Print output
print("Transformer Output :")
print(output)
```
